```python
def count_repite(find_text,file_csv):
    counter = 0
    values = []
    
    with open(file_csv, 'r') as file:
        reader = csv.reader(file)
        for row in reader:
            for i in range(len(row)):
                if find_text in row[i]:
                    counter += 1
                    if i+1 < len(row):
                        values.append(row[i+1])
    return counter, values

find_text = "Could you explain this concept to me in simpler terms?"
file_csv = 'queries_pocasm/for_statistics.csv'
repite, values = count_repite(find_text,file_csv)
print(repite)
print(values)
```

```python
from collections import Counter
# element_count = Counter(values)
# print(element_count)
def sentence_repite_values():
    data = []
    for sentence in df['Query']:
        if sentence != ' ':
            file_csv = 'queries_pocasm/for_statistics.csv'
            repeat, values = count_repite(sentence, file_csv)
            # print(sentence)
            # print(repite)
            # print(values)
            element_count = Counter(values)
            count = element_count.most_common()
            countQuery = 0
            countConversation = 0
            countGeneration = 0
            
            for s in count:
                if s[0] == '  Query':
                    countQuery = s[1]
                elif s[0] == '  Conversation':
                    countConversation = s[1]
                elif s[0] == '  Generation':
                    countGeneration = s[1]
            # print( count)
            currentData = {
                'Sentence': sentence,
                'Repeat': repeat,
                # 'values': values, 
                'Query': countQuery,
                'Conversation': countConversation,
                'Generation': countGeneration
            }
            data.append(currentData)
    # print(data)        
    return data

sentence_repite_values()
```

```python
file_values_repite = 'queries_pocasm/values_repite.csv'
fields = ['Sentence', 'Repeat', 'Query', 'Conversation', 'Generation']
data = sentence_repite_values()
with open(file_values_repite, 'w', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=fields)
    writer.writeheader()
    writer.writerows(data)    
```
